# Задание 1

```python
import pandas as pd
import math
```

```python
x1 = pd.np.linspace(1, 10, 20)
x2 = pd.np.linspace(-20, -1, 50)
x = pd.np.array(pd.np.meshgrid(x1, x2)).T.reshape(-1, 2)
y = pd.np.cos(x[:,0] + x[:,1])

df = pd.DataFrame(pd.np.vstack([x[:,0], x[:,1], y]).swapaxes(0, 1), columns=['x_1', 'x_2', 'y'])
df.to_csv('data.csv', index=False)
# x
df
```

# Задание 2 

```python
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data.csv')

df_1 = df[df['x_2']==x2[0]]
plt.scatter(df_1['x_1'], df_1['y'])
plt.show()

df_2 = df[df['x_1']==x1[0]]
plt.scatter(df_2['x_2'], df_2['y'])
plt.show()
```

# Задание 3

```python
df_agg = df.agg(['mean', 'min', 'max'])
df_agg
```

# Задание 4

```python
df_filtered = df[(df['x_1']<df_agg['x_1']['mean']) | (df['x_2']<df_agg['x_2']['mean'])]
df_filtered.to_csv('data_filtered.csv')
df_filtered
```
